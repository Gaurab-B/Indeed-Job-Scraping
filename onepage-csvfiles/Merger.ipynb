```python
import pandas as pd
from datetime import datetime
```

```python
# Define the paths to the CSV files
path1 = 'AI-Cincinnati-Page(1).csv'
path2 = 'AI-Cincinnati-Page(2).csv'
path3 = 'AI-Cincinnati-Page(3).csv'
path4 = 'AI-Cincinnati-Page(4).csv'
path5 = 'AI-Cincinnati-Page(5).csv'

# Read each CSV file into a Pandas dataframe
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df3 = pd.read_csv(path3)
df4 = pd.read_csv(path4)
df5 = pd.read_csv(path5)
```

```python
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
df['Scraped_On'] = datetime.today().strftime('%Y-%m-%d')
df
```

```python
df.to_csv("Artificial-Intelligence-Cincinnati(1-5).csv")
```
