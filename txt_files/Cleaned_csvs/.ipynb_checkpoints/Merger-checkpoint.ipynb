```python
import os
import pandas as pd
```

```python
def merge_csv_files(folder_path, output_file):
    # List all CSV files in the folder
    file_list = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    # Sort files by their numerical page number
    file_list.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))

    # Initialize an empty list to hold DataFrames
    dfs = []

    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        dfs.append(df)
        print(f'Reading {file_name}')

    # Concatenate all DataFrames into one
    combined_df = pd.concat(dfs, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    combined_df.to_csv(output_file, index=False)
    print(f'Saved combined CSV to {output_file}')

# Example usage
folder_path = '.'
output_file = 'merged_output.csv'
merge_csv_files(folder_path, output_file)
```

```python
def csv_to_readme(csv_file_path, readme_file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Open the README.md file in write mode with UTF-8 encoding
    with open(readme_file_path, 'w', encoding='utf-8') as readme_file:
        # Write a header to the README.md file
        readme_file.write('# Merged CSV Data\n\n')
        
        # Write the DataFrame contents as a table
        readme_file.write(df.to_markdown(index=False))

    print(f'Created {readme_file_path} with the contents of {csv_file_path}')

# Example usage
csv_file_path = 'merged_output.csv'
readme_file_path = 'merded_output.md'
csv_to_readme(csv_file_path, readme_file_path)
```
